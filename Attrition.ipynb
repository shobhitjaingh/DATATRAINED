```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(sampling_strategy="minority")
x_new,y_new=smote.fit_resample(x,y)
y_new.value_counts()
```

# FINDING BEST RANDOM STATE

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import roc_auc_score

from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
maxacc=0
maxRN=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=.33,random_state=i)
    Lr=LogisticRegression()
    Lr.fit(x_train,y_train)
    pred=Lr.predict(x_test)
    r2=accuracy_score(y_test,pred)
    if r2>maxacc:
        maxacc=r2
        maxRN=i
print("BEST accuracy score is ",maxacc,"on Random_State",maxRN) 
```

We found our best random state as 122

```python
x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=.33,random_state=122)
```

LOGISTIC REGRESSION

```python
Lr=LogisticRegression()
Lr.fit(x_train,y_train)
predlr=Lr.predict(x_test)
print("Accuracy Score", accuracy_score(y_test,predlr))
print("Confusion Matrix", confusion_matrix(y_test,predlr))
print("Classification Report",classification_report(y_test,predlr))
print("roc_auc_score",roc_auc_score(y_test,predlr))   
```

DECISION TREE CLASSIFIIER

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
preddt=dt.predict(x_test)
print("Accuracy Score", accuracy_score(y_test,preddt))
print("Confusion Matrix", confusion_matrix(y_test,preddt))
print("Classification Report",classification_report(y_test,preddt))
    
```

SUPPORT VECTOR CLASSIFIER

```python
from sklearn.svm import SVC
sv=SVC()
x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=.33,random_state=55)
sv.fit(x_train,y_train)
predsv=sv.predict(x_test)
print("Accuracy Score", accuracy_score(y_test,predsv))
print("Confusion Matrix", confusion_matrix(y_test,predsv))
print("Classification Report",classification_report(y_test,predsv))
print(roc_auc_score(y_test,predsv))
print()  
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
kn.fit(x_train,y_train)
predkn=kn.predict(x_test)
print("Accuracy Score", accuracy_score(y_test,predkn))
print("Confusion Matrix", confusion_matrix(y_test,predkn))
print("Classification Report",classification_report(y_test,predkn))
print()  
```

# Ensembling Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print("Accuracy Score", accuracy_score(y_test,predrf))
print("Confusion Matrix", confusion_matrix(y_test,predrf))
print("Classification Report",classification_report(y_test,predrf))
print() 
```

# Cross Validation

```python
models = [LogisticRegression(), DecisionTreeClassifier(), SVC(),RandomForestClassifier(),KNeighborsClassifier()]
model_names = [type(x).__name__ for x in models]

std_score = []
for m in models:
    std_score.append(cross_val_score(estimator = m,
                                 X = x_new,
                                 y = y_new,
                                 scoring = 'accuracy',
                                 cv = 5).mean())
    
pd.Series(data = std_score, index = model_names)
```

By comparing the accuracy of all models between their model accuracy and cross validation score we choose leat iffernce model i.e SVC As best i.e 87

# HYPERTUNING OF MODEL

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters= {'C': [0.1, 2.0],
             'kernel':  ['linear', 'poly', 'rbf', 'sigmoid'],
             'degree':  [2, 5, 1],
             'gamma': ['auto', 'scale'],
             'max_iter': [-1, 100, 1]}
```

```python
GCV=GridSearchCV(SVC(),parameters,cv=5)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
Final_model=SVC(C=.1,degree=5,gamma='scale',kernel='poly',max_iter=-1)
Final_model.fit(x_train,y_train)
pred=Final_model.predict(x_test)
acc=accuracy_score(y_test,pred)
roc_auc=roc_auc_score(y_test,pred)
print(acc)
print(roc_auc)
```
